```python
# 연습 문제
# urllib 임포트
from urllib.request import urlopen
from bs4 import BeautifulSoup

# 다음 뉴스에서 받아오기
req = urlopen("https://news.daum.net/")
print("응답 코드: ", req.getcode())

if req.getcode() == 200:
    # html 받아오기
    html = req.read()
    html = html.decode("utf-8")
    print("SUCCESS")
else:
    print("HTTP-ERROR")

# 문서 정보 확인
print("문서 제목: ", soup.title.text)
print("")

# text -> html dom으로 변경
soup = BeautifulSoup(html, "html.parser")
# print(soup.prettify())

# Daum 뉴스 상단 div#cSub인 뉴스 수집
articles = soup.select("#cSub")[0]
# print(articles)

# div#cSub 의 자손인 a.link_txt를 검색
links = articles.select("a.link_txt")

# 기사 목록을 리스트화
news_list = []

for link in links:
    news = {
        "link": link["href"],
        "title": link.text.strip()
    }
    news_list.append(news)

# print(news_list)
print("[다음 뉴스 메인 기사]")

for news in news_list:
    print(f"{news['title']} - {news['link']}")
```

```python
# 네이버 영화 순위를 수집
req = urlopen("https://movie.naver.com/movie/running/current.nhn")
html_str = req.read().decode('utf-8')
# print(html_str)
```

```python
# DOM 변환
soup = BeautifulSoup(html_str, "html.parser")
# print(soup.body)
```

```python
# ul.list_detail_t1
# class=lst_detail_t1인 ul의 자식인 li태그들
current_movies = soup.select("ul.lst_detail_t1 > li")  
# print(current_movies)

# 영화의 세부 정보 검색
movie_list = []

for movie in current_movies:
    titles = movie.select("dl.lst_dsc > .tit > a")[0]
    
    info = {"title": titles.text.strip(), "link": titles["href"]}
    movie_list.append(info)

# print(movie_list)
print("네이버 영화 순위")

for rank, info in enumerate(movie_list):
    print(f"{rank + 1}위 - {info['title']}: {info['link']}")
```

```python
# 이미지 다운로드
# ul.lst_detail_t1 > li -> img
# img 태그의 속성 src, alt(대체 텍스트) 확인
# 저장 디렉토리 확인
import os
import re  # 정규 표현식
from urllib.request import urlretrieve

# 디렉토리가 없으면 디렉토리 생성
if not os.path.exists("d:/data-science/crawl"):
    os.makedirs("d:/data-science/crawl")

for movie in current_movies:
    img_tag = movie.select("img")[0]
#     print("포스터: ", img_tag)
#     print("src: ", img_tag['src'])
#     print("alt: ", img_tag['alt'])
    # 정규식을 사용하여 파일명으로 사용할 수 없는 문자 제거
    title = img_tag['alt']
    title = re.sub(r"[\/:*?<>|.]", " -", title.strip())
#     print("정제된 제목: ", title)
    src = img_tag['src']  # 이미지 주소
    target = f"d:/data-science/crawl/{title}.jpg"
    
    # 이미지 다운로드: urlretrieve
    urlretrieve(src, target)
    print(".", end="")
else:
    print("Download Complete!")
```

```python
# 연습 문제
# https://movie.naver.com/movie/running/premovie.nhn
# 개봉 예정 영화 정보를 수집
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re  # 정규 표현식

req = urlopen("https://movie.naver.com/movie/running/premovie.nhn")
html = req.read().decode("utf-8")
soup = BeautifulSoup(html, "html.parser")

# 영화 세부 정보 리스트
current_movies = soup.select("ul.lst_detail_t1 > li")
movie_list = []

for movie in current_movies:
    title = movie.select("dl.lst_dsc > dt.tit > a")[0].text
    title = re.sub(r"[\/:*?<>|.]", " -", title.strip())   # 제목 정제
    rating = movie.select("dl.lst_dsc > dt.tit > span")
    rating = rating[0].text if len(rating) else "미정"   # rating 이 없는 항목의 경우, 임의의 문자열 "미정"을 저장
    url = movie.select("dl.lst_dsc > dt.tit > a")[0]['href']
    
    info = {
        "title": title,
        "rating": rating,
        "url": url
    }
    
    movie_list.append(info)

for movie in movie_list:
    print(f"영화 제목: {movie['title']}")
    print(f"관람 등급: {movie['rating']}")
    print(f"영화 url : {movie['url']}")
    print("")
```
