```python
# 데이터 셋의 확인
from sklearn.datasets import load_iris
iris_dataset = load_iris()
type(iris_dataset)
```

```python
# Bunch의 내용 확인
print(iris_dataset.keys())
```

```python
# 데이터 셋의 설명서 확인
print(iris_dataset['DESCR'])
```

```python
# 분류해야 할 target의 이름
print("Target names: ", iris_dataset['target_names'])
```

```python
# target의 확인
print("Target: ", iris_dataset['target'])
```

```python
# 학습 데이터를 DataFrame으로 확인
import pandas as pd
iris_df = pd.DataFrame(iris_dataset['data'], columns = iris_dataset['feature_names'])
iris_df
```

```python
tnames = [iris_dataset['target_names'][i] for i in iris_dataset['target']]
target_df = pd.DataFrame(zip(iris_dataset['target'], tnames), columns = ['target', 'target_names'])
target_df
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (10, 10))
plt.scatter(iris_df['petal length (cm)'], iris_df['petal width (cm)'], c = target_df['target'])
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
```

```python
# 산점도 매트릭스
pd.plotting.scatter_matrix(iris_df, 
                           c = iris_dataset['target'], 
                           figsize = (10, 10),
                           marker = 'o',
                           hist_kwds = {"bins": 20},
                           s = 60, alpha = 8)
plt.show()
```

```python
# 학습 데이터와 테스트 데이터 분할
from sklearn.model_selection import train_test_split
# 분할
X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], 
                                                    iris_dataset['target'],
                                                    random_state = 0)
```

```python
print("Size of X_train:", X_train.shape)
print("Size of X_test:", X_test.shape)
print("Size of y_train:", y_train.shape)
print("Size of y_test:", y_test.shape)
```

```python
# 학습 진행
# kNN 알고리즘 사용
from sklearn.neighbors import KNeighborsClassifier
# 모델
knn = KNeighborsClassifier(n_neighbors = 3)   # 이웃 3개를 기준으로 분류
```

```python
knn.fit(X_train, y_train)   # 지도 학습이니까 학습 데이터화 정답(라벨)
```

```python
# 새로운 데이터를 이용한 품종의 예측
import numpy as np

X_new = np.array([[5, 2.9, 1, 0.2]])

prediction = knn.predict(X_new)
print("예측값: ", prediction)
print("예측한 품종: ", iris_dataset['target_names'][prediction])
```

```python
# 평가(Evaluate)
# X_test(테스트 데이터 셋)으로 예측
y_pred = knn.predict(X_test)
print("테스트 세트에 대한 예측값: ", y_pred)
print("테스트 세트 라벨(정답): ", y_test)
```

```python
# 정확도 판단
print("테스트 세트의 정확도: {:.2f}".format(np.mean(y_pred == y_test)))
```

```python
# 정확도 판단(score)
print("테스트 세트의 정확도: {:.2f}".format(knn.score(X_test, y_test)))
```
